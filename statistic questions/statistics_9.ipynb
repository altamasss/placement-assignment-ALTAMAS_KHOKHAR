 Given the score of students in multiple exams

 Test the hypothesis that the mean scores of all the students are the same. If not, name the student with the highest score.

```python
import scipy.stats as stats
# Given data
exam1 = [85,70,90,75,95]
exam2 = [90,80,85,70,92] 
final_exam = [92,85,88,75,96]

# Performing one-way ANOVA
f_statistic, p_value = stats.f_oneway(exam1,exam2,final_exam)

# taking significance level as 0.05
alpha = 0.05

# Checking for rejection of  null hypothesis
if p_value < alpha:
    print("The mean scores of the students are not the same.")
    
    
    all_scores = exam1 + exam2 + final_exam
    highest_score = max(all_scores)
     
        
    student_with_highest_score = [i + 1 for i, score in enumerate(all_scores) if score == highest_score]
    print(f"The student with the highest score is: {student_with_highest_score}")
    
else:
    print("The mean scores of the students are the same.")
```
