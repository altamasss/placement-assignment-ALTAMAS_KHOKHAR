# Question no.8 :- 

The anxiety levels of 10 participants were measured before and after a new therapy. The scores are not normally distributed. Use the Wilcoxon signed-rank test to test whether the therapy had a significant effect on anxiety levels. The data is given below: Participant Before therapy After therapy Difference

8.png

```python
import scipy.stats as stats

anxiety_before_therapy = [10,8,12,15,6,9,11,7,14,10] 

anxiety_after_therapy = [7,6,10,12,5,8,9,6,12,8]

# Performing wilcoxon signed-rank test
statistic,p_value = stats.wilcoxon(anxiety_before_therapy, anxiety_after_therapy)
alpha = 0.05

if p_value < alpha:
    print("Yes,the Therapy had a significant effect on the Anxiety levels.")
else:
    print("No,the Therapy had no significant effect on the Anxiety levels.")
```
